```python
import matplotlib.pyplot as plt
```

```python
#Basic pie graph

sizes=[3,4,6,2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
# some modifications

labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","purple","red","pink"]
plt.title("Split among classes")
plt.pie(sizes,colors=colors,labels=labels)
plt.axis("equal")
plt.show()
```

```python


labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","purple","red","pink"]
plt.title("Split among classes")
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%")    # will show upto 2 decimal points 
plt.axis("equal")
plt.show()
```

```python
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","purple","red","pink"]

explode=[0.1,0,0,0.3]
plt.title("Split among classes")
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=100) #upto 2 decimal points 
plt.axis("equal")
plt.show()
```
