```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="black",marker="p")
plt.show()
```

```python
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="green",linewidth=5)
plt.ylabel("X^3")
plt.xlabel("x")
plt.title("Matplotlib demo")
plt.show()
```

```python
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="blue",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",linewidth=5,label="x^2")
plt.ylabel("y")
plt.xlabel("x")
plt.legend() # to give specific label to each line
plt.title("Matplotlib demo")
plt.show()
```

```python
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="blue",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",linewidth=5,label="x^2")

plt.axis([0,10,0,100]) # it means take x till 10 and y till 100
plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.legend() # to give specific label to each line
plt.title("Matplotlib demo")
plt.show()
```

```python
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="black",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",linewidth=5,label="x^2")

plt.axis([0,10,0,200]) # it means take x till 10 and y till 200
plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2,80,"Text",fontsize=16)

plt.legend() # to give specific label to each line
plt.title("Matplotlib demo")
plt.show()
```
