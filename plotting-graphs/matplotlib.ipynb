```python
x=[1,2,3]
y=[2,4,6]

#it shows only points
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

#connected with line
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

x2=[3,4,5]
y2=[2,4,6]

# to choose our own colour
plt.plot(x2,y2,"g-.")

plt.plot(x,y)
plt.show()
```
